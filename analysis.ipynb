```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Features (X) and Labels (y)
X = df['job_title']
y = df['experience_level']
```

```python
# Convert job titles to TF-IDF vectors
vectorizer = TfidfVectorizer()
X_vectorized = vectorizer.fit_transform(X)
```

```python
# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X_vectorized, y, test_size=0.2, random_state=42
)
```

```python
# Train a Logistic Regression model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Predict on test set
y_pred = model.predict(X_test)

# Print results
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# Features (input): job titles
X = df['job_title']

# Labels (target): experience level (EN, MI, SE, EX)
y = df['experience_level']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X_vectorized = vectorizer.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_vectorized, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = model.predict(X_test)

print(" Accuracy:", accuracy_score(y_test, y_pred))
print("\n Classification Report:\n", classification_report(y_test, y_pred))
```

```python
job_input = "Lead Data Engineer"
vec_input = vectorizer.transform([job_input])
prediction = model.predict(vec_input)
print("Predicted Experience Level:", prediction[0])
```

```python
#  Predict experience level from a new job title
job_input = "Lead Data Engineer"  #  You can change this

# Transform the input using the same vectorizer
vec_input = vectorizer.transform([job_input])

# Predict using the trained model
prediction = model.predict(vec_input)

print(f" Job Title: {job_input}")
print(f" Predicted Experience Level: {prediction[0]}")
```
